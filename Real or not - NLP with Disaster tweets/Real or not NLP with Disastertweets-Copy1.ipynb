```python
#input_d = train_data[['keyword','location','text']].values.tolist()
input_d = train_data['text']
```

```python
input_data = input_d.tolist()
input_data
```

```python
target = train_data['target'].tolist()
target
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
```

```python
vectorizer
```

```python
x = vectorizer.fit_transform(input_data)
x
```

```python
feature_names = vectorizer.get_feature_names()
feature_names
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
stop_words
```

```python
finallist = [word for word in feature_names if word not in stop_words]
finallist[300]
```

```python
tf_features = pd.DataFrame(x.toarray(),columns=vectorizer.get_feature_names())
```

```python
tf_features[0:1]
```

```python
vectorizer = CountVectorizer(max_df =.85, stop_words=stop_words,max_features=10000,ngram_range=(2,5),token_pattern= '[a-zA-Z\-][a-zA-Z\-]{2,}')
#max_df = any word that appears 85%, min_df= any wordwhich appears very less
```

```python
X = vectorizer.fit_transform(input_data)
```

```python
bow_tf_df = pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names())
```

```python
ip = bow_tf_df
op = target
```

```python
ip.shape
```

```python
ip[0:1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(ip, op, test_size=.3 )
```

```python
from sklearn import tree 
ctree = tree.DecisionTreeClassifier()
ctree.fit(x_train, y_train)
y_pred = ctree.predict(x_test)

from sklearn import metrics
print(metrics.accuracy_score(y_test, y_pred))
```
