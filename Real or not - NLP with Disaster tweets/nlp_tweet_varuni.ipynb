```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_df =.85, stop_words=stop_words,max_features=2500,ngram_range=(2,2),token_pattern= '[a-zA-Z\-][a-zA-Z\-]{2,}')
#max_df = any word that appears 85%, min_df= any wordwhich appears very less
```

```python
x = vectorizer.fit_transform(train_tweet.iloc[:,3])
```

```python
bow_tf_df = pd.DataFrame(x.toarray(),columns=vectorizer.get_feature_names())
```

```python
x = bow_tf_df
y = train_tweet.iloc[:,4]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.3 )
```

```python
from sklearn import tree 
ctree = tree.DecisionTreeClassifier()
ctree.fit(x_train, y_train)
y_pred = ctree.predict(x_test)

from sklearn import metrics
print(metrics.accuracy_score(y_test, y_pred))
```
